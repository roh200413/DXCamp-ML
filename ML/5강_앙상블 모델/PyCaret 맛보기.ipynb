```python
# 공통 모듈 임포트
import sklearn
import numpy as np
import pandas as pd

# 노트북의 실행 결과가 동일하도록
np.random.seed(42)

# 깔금한 그래프 출력을 위해
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
```

```python
# 데이터 불러오기 및 판다스 데이터프레임 객체에 담기: pd.read_csv()
data = pd.read_csv('./drive/MyDrive/06. DX 캠프 코드/5강_앙상블 모델/bike-sharing-demand/train.csv')
```

```python
from pycaret.regression import *

# 데이터 분석 워크플로우 자동화
setup_clf = setup(data=data, target='count')
```

```python
# 회귀와 관련된 대부분의 모델 생성 및 데이터 학습
best = compare_models(exclude = ['ransac'])
```

```python
# 특정 모델 가져오기
lr = create_model('lr')
tuned_lr = tune_model(lr)
```

```python
predictions = predict_model(tuned_lr, data = data.drop('count', axis = 1))
predictions
```

```python
predictions['Label']
```

```python
# 여러개의 분류기로 앙상블 모델 생성
best_models = compare_models(sort='MSE', n_select=5, fold=5)
```

```python
blended = blend_models(estimator_list = best_models, fold = 5)
predictions = predict_model(blended, data = data.drop('count', axis = 1))
predictions
```
