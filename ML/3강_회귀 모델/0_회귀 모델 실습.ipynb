# 3. 다항 회귀

```python
bias = 1
beta = np.array([2,3]).reshape(2, 1)
noise = np.random.randn(100, 1)
```

```python
beta
```

```python
np.random.seed(42)
X = np.random.randn(100, 1)
X_poly = np.hstack([X, X**2])
```

```python
X[:10]
```

```python
X_poly[:10]
```

```python
y = bias + X_poly.dot(beta)
y_with_noise = y + noise
```

```python
plt.scatter(X, y_with_noise)
```

```python
model = LinearRegression()
model.fit(X_poly, y_with_noise)
```

```python
print(f'편향: {model.intercept_}')
print(f'회귀계수: {model.coef_}')
```

```python
X_test = np.array([i for i in np.linspace(-3,4,100)]).reshape(100,1)

X_poly = np.hstack([X_test, X_test**2])
```

```python
pred = model.predict(X_poly)
```

```python
pred[:5]
```

```python
plt.scatter(X, y_with_noise)
plt.plot(X_test, pred, color='green')
```

# 4. 릿지 회귀, 라쏘 회귀

image.png

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.preprocessing import PolynomialFeatures
```

```python
X = np.array([2,4,6,8,10,12,14,16])
y = np.array([4,2,8,6,12,10,16,18])
X = X.reshape((8,1))
poly_features = PolynomialFeatures(degree=5, include_bias=False)
X_poly2 = poly_features.fit_transform(X)

def ridgePlot(alpha):
  ridge = Ridge(alpha=alpha).fit(X_poly2, y)
  X_test = np.array([i for i in np.linspace(-5,25,100)]).reshape(100,1)
  X_poly3 = poly_features.fit_transform(X_test)
  pred = ridge.predict(X_poly3)
  plt.scatter(X, y)
  plt.plot(X_test, pred, color='green')

ridgePlot(0.1)
```

```python
ridgePlot(100)
```

```python
X = np.array([2,4,6,8,10,12,14,16])
y = np.array([4,2,8,6,12,10,16,18])
X = X.reshape((8,1))
poly_features = PolynomialFeatures(degree=10, include_bias=False)
X_poly2 = poly_features.fit_transform(X)

def lassoPlot(alpha):
  lasso = Lasso(alpha=alpha).fit(X_poly2, y)
  X_test = np.array([i for i in np.linspace(-2,20,100)]).reshape(100,1)
  X_poly3 = poly_features.fit_transform(X_test)
  pred = lasso.predict(X_poly3)
  plt.scatter(X, y)
  plt.plot(X_test, pred, color='green')

lassoPlot(0.1)
```

```python
lassoPlot(10)
```
