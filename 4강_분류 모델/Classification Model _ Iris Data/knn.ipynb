## k 값에 따른 결정 경계

k 값에 따른 knn의 결정경계를 그려봅니다.  
k 가 작을수록 overfitting이 k가 클수록 underfitting이 됩니다.

```python
# 시각화 데이터
x_min, x_max = data[:, 0].min() - 1, data[:, 0].max() + 1
y_min, y_max = data[:, 1].min() - 1, data[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

from sklearn.neighbors import KNeighborsClassifier

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(20, 10))
for idx, n in enumerate(range(1, 12, 2)):
    # knn 생성 및 학습
    knn = KNeighborsClassifier(n_neighbors=n).fit(data, target)

    # 시각회 데이터 예측
    Z = knn.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    
    ax = axes[idx//3, idx%3]

    # 영역 표시
    ax.contourf(xx, yy, Z)

    # 데이터 표시    
    ax.scatter(data[:, 0], data[:, 1], c=target, alpha=1.0, edgecolor="black", s = 80)
    ax.set_title(f"{n} Nearest Neighbors")
    #ax.set_xlim(xx.min(), xx.max())
    #ax.set_ylim(yy.min(), yy.max())
    #ax.set_xlabel(iris.feature_names[0])
    #ax.set_ylabel(iris.feature_names[1])
```

## 나의 가장 가까운 이웃은?

KNN의 거리의 종류는 p를 통해서 바꿀 수 있습니다.
- p=1 => 맨해튼 거리
- p=2 => 유클리드 거리

image.png

### 유클리드 거리(Euclidean Distance)

```python
train_data, train_target = data[:-1], target[:-1]
test_data = data[-1:]
```

```python
from sklearn.neighbors import KNeighborsClassifier

euclid_knn = KNeighborsClassifier(n_neighbors=10)
euclid_knn.fit(train_data, train_target)
```

```python
euclid_knn.kneighbors(
    test_data, n_neighbors=1, return_distance=False
).ravel()
```

```python
euclid_neighbors_idx = euclid_knn.kneighbors(
    test_data, n_neighbors=10, return_distance=False
).ravel()

euclid_neighbors_idx
```

```python
euclid_neighbors = train_data[euclid_neighbors_idx]
euclid_neighbors_label = train_target[euclid_neighbors_idx]
```

```python
test_data
```

```python
euclid_neighbors
```

```python
euclid_neighbors_label
```

```python
euclid_knn.predict(test_data)
```

```python
euclid_knn.predict_proba(test_data)
```

```python
plt.figure(figsize=(15, 8))
plt.scatter(train_data[:, 0], train_data[:, 1], c=train_target, s=200)
plt.scatter(test_data[0, 0], test_data[0, 1], marker="X", s=500)
plt.scatter(euclid_neighbors[:, 0], euclid_neighbors[:, 1], c=euclid_neighbors_label, edgecolors="red", linewidth=5, s=200)
```

### 맨하튼 거리(Manhattan Distance)

```python
manhattan_knn = KNeighborsClassifier(n_neighbors=10, p=1)
manhattan_knn.fit(train_data, train_target)
```

```python
manhattan_neighbors_idx = manhattan_knn.kneighbors(
    test_data, n_neighbors=10, return_distance=False
).ravel()
manhattan_neighbors = train_data[manhattan_neighbors_idx]
manhattan_neighbors_label = train_target[manhattan_neighbors_idx]
```

```python
manhattan_neighbors
```

```python
manhattan_neighbors_label
```

```python
manhattan_knn.predict_proba(test_data)
```

```python
plt.figure(figsize=(15, 8))
plt.scatter(train_data[:, 0], train_data[:, 1], c=train_target, s=200)
plt.scatter(test_data[0, 0], test_data[0, 1], marker="X", s=500)
plt.scatter(manhattan_neighbors[:, 0], manhattan_neighbors[:, 1], c=manhattan_neighbors_label, edgecolors="red", s=200, linewidth=5)
```

### 거리 비교

```python
print(f"유클리디언 거리: {euclid_neighbors_idx}")
print(f"맨하튼 거리: {manhattan_neighbors_idx}")
```

```python
set(euclid_neighbors_idx) - set(manhattan_neighbors_idx)
```

```python
set(manhattan_neighbors_idx) - set(euclid_neighbors_idx)
```

```python
#결과가 다른 데이터
diff_neighbors_idx = list(set(euclid_neighbors_idx) - set(manhattan_neighbors_idx))
diff_neighbors_idx.extend(list(set(manhattan_neighbors_idx) - set(euclid_neighbors_idx)))
print(diff_neighbors_idx)

diff_neighbors = train_data[diff_neighbors_idx]
diff_neighbors_label = train_target[diff_neighbors_idx]
```

```python
#결과가 겹치는 데이터
same_neighbors_idx = list(set(euclid_neighbors_idx) & set(manhattan_neighbors_idx))
print(same_neighbors_idx)

same_neighbors = train_data[same_neighbors_idx]
same_neighbors_label = train_target[same_neighbors_idx]
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
fig.set_size_inches(20, 6)

# 맨하든 거리
ax1.scatter(train_data[:, 0], train_data[:, 1], c=train_target, s=100)
ax1.scatter(test_data[0, 0], test_data[0, 1], marker="X", s=200)
ax1.scatter(euclid_neighbors[:, 0], euclid_neighbors[:, 1], c=euclid_neighbors_label, edgecolors="red", linewidth=4, s=100)

# 유클리드 거리
ax2.scatter(train_data[:, 0], train_data[:, 1], c=train_target, s=100)
ax2.scatter(test_data[0, 0], test_data[0, 1], marker="X", s=200)
ax2.scatter(manhattan_neighbors[:, 0], manhattan_neighbors[:, 1], c=manhattan_neighbors_label, edgecolors="red", s=100, linewidth=4)

# 각 거리 비교: 파란색 데이터 = 겹치는 데이터, 빨간색 데이터 = 겹치지 않는 데이터
fig, ax3 = plt.subplots(nrows=1, ncols=1)
fig.set_size_inches(8, 5)
ax3.scatter(train_data[:, 0], train_data[:, 1], c=train_target, s=100)
ax3.scatter(test_data[0, 0], test_data[0, 1], marker="X", s=300)
ax3.scatter(diff_neighbors[:, 0], diff_neighbors[:, 1], c=diff_neighbors_label, edgecolors="red", s=100, linewidth=5)
ax3.scatter(same_neighbors[:, 0], same_neighbors[:, 1], c=same_neighbors_label, edgecolors="blue", s=100, linewidth=5)
```
